```python
#%% 11. Escribir un programa que reciba una cadena de texto y reporte:
#a) cuántas letras vocales en mayúscula se entraron;
#b) cuántas letras con tilde se entraron (minúsculas y mayúsculas);
#c) cuántos dígitos se entraron,
#d) cuántos espacios se entraron;
#e) cuántas palabras reservadas se entraron.
    
# función para detectar si un caracter es una vocal mayúscula 
def vocalmayus(carac):
    
    return  carac == 'A' or carac == 'E' or carac == 'I' or carac == 'O' or\
            carac == 'U' or carac == 'Á' or carac == 'É' or carac == 'Í' or\
            carac == 'Ó' or carac == 'Ú'
# función que determina si un caracter tiene una tilde (solo se buscará 
#para las vocales)
def tilde(carac2):
    
    return carac2.upper() == 'Á' or carac2.upper() == 'É' \
           or carac2.upper() == 'Í' or carac2.upper() == 'Ó' \
           or carac2.upper() == 'Ú'  
# ingreso de cadena a estudiar
cadena = input('Ingrese un texto: ')


cuantos_digitos = 0 # variable del número de dígitos en la cadena
cuantos_espacios = 0 # variable del número de espacios en la cadena
cuantas_vocal_mayus = 0 # variable del número de vocales en mayúsculas
cuantas_letras_tilde = 0 # variable del número de letras con tilde

for i in range(0, len(cadena)):
    
    if cadena[i].isspace(): # encuentra un espacio
        cuantos_espacios += 1
    elif cadena[i].isdigit(): # encuentra un dígito
        cuantos_digitos += 1
    elif vocalmayus(cadena[i]): # si una letra es vocal en mayúscula
        cuantas_vocal_mayus += 1
    elif tilde(cadena[i]): # encuentra una vocal con tilde 
                                 # (mayúscula o minúscula)
        cuantas_letras_tilde += 1
print(f'La cadena de texto "{cadena}" tiene:')
print(f'a) {cuantas_vocal_mayus} vocales en mayúscula')
print(f'b) {cuantas_letras_tilde} vocales con tilde (mayúscula o minúscula)')
print(f'c) {cuantos_digitos} dígitos')
print(f'd) {cuantos_espacios} espacios')
```

```python
#%% 15 Leer una lista de números enteros y enumerar cuantos elementos se 
#repiten exactamente dos veces. Utilizar para tal fin el método count().
lista=[] #Elemento vacio para almacenar la lista ingresada por el usuario
repetidos=[]  #Lista vacia para guardar los numeros repetidos 
cantidad=int(input('Ingrese la cantidad de numeros de la lista: '))#Cantidad 
                                                                   #de datos 
                                                                  #a recolectar
longrep=int(input('Ingrese la cantidad de veces que se repite el numero\
en la lista: ')) #Cantidad de veces que se repite el numero en la lista
while longrep>cantidad: #mientras la cantidad de numeros a buscar sea mayor a
    #la longitud de la lista se generara un error 
    print('ERROR,la cantidad de veces que se repite el numero debe ser menor a\
la longitud total de la lista')
    cantidad=int(input('Ingrese la cantidad de numeros de la lista: '))
    longrep=int(input('Ingrese la cantidad de veces que se repite el numero en\
    la lista: '))
    
    
for i in range(cantidad): #ciclo para la recoleccion
    lista.append(float(input(f'Escriba el numero {i+1} de la lista: ')))#
                                              #Se añade el elemnto i a la lista

#ANALISIS DE LA LISTA INGRESADA

for i in range(cantidad): #Ciclo para el recorrido en la lista recolectada
    sentencia=lista[i] in repetidos  #En sentencia se analiza si ya se conto el
    # valor de la lista por ejemplo en el numero 2254748 para la primera vuelta 
    #se buscara si el 2 esta almacenado en la lista de repetidos, dando como 
    #resultado falso                                     
    if sentencia==False: #Para el resultado falso se contara la cantidad de 
                 #veces que esta el valor 2 en la lista, dando como resultado 2
        a=lista.count(lista[i]) #En a se almacena el numero de veces que esta 
                                #el numero en la lista
        if a==longrep: #Si a es dos significa que esta exactamente 2 veces en 
                       #la lista y cumple el requisito
            repetidos.append(lista[i]) #Se agrega el numero que cumple la 
                                       #condicion en la lista de repetidos
#Para la segunda vuelta del ciclo se confirmara si 2 ya se guardo en la lista
# de repetidos, dando como resultado True Dado que ese valor de la lista ya se 
#conto se sigue con el siguiente elemtno de la lista
listordenada=sorted(repetidos)    #Ordenamiento de menor a mayor
                                  # de los numeros repetidos
print(f'los siguientes numeros se encontraron exactamente {longrep} veces en \
la lista: {listordenada} ' )

```

```python
#%% 16 Leer una lista de números ya ordenados de forma ascendente y verificar
 #que dicha lista está ordenada.
lista=[] #Lista vacia para recolectar los datos ingresados por el usuario
añadir=[] #Lista vacia para almacenar los datos a añdir a la lista original
cantidad=int(input('ingrese la cantidad de numeros de la lista: '))#Cantidad de 
                                                           #datos a recolectar
for i in range(cantidad):  #ciclo para recoleccion de datos
    lista.append(float(input(f'Escriba el numero {i+1} de la \
    lista ordenada: ')))   #almacenamiento de datos lsita orignial

comprobacion=lista==sorted(lista) #Parametro de Comparacion de las listas
                                  # organizada e ingresada
if comprobacion== True: #Si son iguales se retorna
                        # True e indica que esta organizada
    print('la cadena esta organizada')
else:
    print('la cadena no esta organizada')

insert=int(input('ingrese la cantidad de numeros a insertar: ')) #Cantidad de 
                                                               #numeros a 
                                                #insertar en  la lista orignial
for i in range(insert): #ciclo para la recoleccion de los valores a insertar
    añadir.append(float(input(f'Escriba el numero {i+1} a \
   insertar en la lista: ')))                    #almacenamiento de datos nuevos

# Ciclo para ordenar la lista
listafin=lista+añadir #la suma de ambas listas dara la lista final a organizar

for recorrido in range(1,len(listafin)): #ciclo para el recorrido de la lista 
                                #nueva y ordenamiento la lista de menor a mayor
    for posicion in range(len(listafin)-recorrido): 
        if listafin[posicion]>listafin[posicion+1]: #compara el elemento actual
       #de la lista con el siguiente valor si es mayor se cambiaran de posicion
            temp=listafin[posicion]  #valor temporal para almacenar el valor 
                                     #orignial
            listafin[posicion]=listafin[posicion+1] #Se remplaza la posicion 
                                #actual con el numero que le seguia en la lista
            listafin[posicion+1]=temp #Con la copia temporal se ubica el
            # numero, logrando asi el cambio de posicion
print('Su lista final es: ',listafin)
```

```python
#%% 17 Leer una lista y determinar cuál es el segundo elemento más grande 
#de dicha lista. No utilice ningún tipo de algoritmo de ordenamiento 
#para realizar este ejercicio
lista=[] #Lista vacia para el almacenamiento de datos
pibote=[] #almacenamiento de los pibotes encontrados (numeros mayores 
          #hasta el momento)
cantidad=int(input('ingrese la cantidad de numeros de la lista: '))#Cantidad de 
                                                            #datos a recolectar
for i in range(cantidad): #ciclo para recoleccion de valores de la lista
    lista.append(float(input(f'Escriba el numero {i+1}\
    de la lista ordenada: ')))
parar=lista[0] #se toma el primer elemento de la lista como pibote y se 
               #genera la condicion parar

for i in range(len(lista)):#Ciclo para recorrer la lista ingresada
    if lista[i]>parar:     #compara el pibote con el numero i de la lista
        parar=lista[i]     # si el valor de la lista es mayor al pibote actual
                           #lo sustituye hasta acabar de recorrer toda la lista
        pibote.append(parar) #Se almacena el pibote actual
print(f'El segundo mayor numero de la lista es {pibote[len(pibote)-2]} \
y se encuetra en la posicion numero {lista.index(pibote[len(pibote)-2])+1} \
de la lista') #Se toma el penultimo pibote guardado
```

```python
#%% 18 Leer una matriz cuadrada A, obtener su matriz triangular superior
#asociada U y determinar qué fila de U contiene la mayor cantidad de unos.
matriz=[] #crea la variable matriz
filas=int(input('ingrese el numeor de filas: ')) #Numero de filas de la matriz
columnas=int(input('ingrese el numeor de columnas: ')) #Numero de columnas 

while filas != columnas: #Como requisito se debe ingresar una matriz cuadrada
    print('ERROR infgrese una matriz cuadrada')
    filas=int(input('ingrese el numeor de filas: ')) #Se ingresa nuevaente las 
                                     #dimensiones ahsta que cumpla la condicion
    columnas=int(input('ingrese el numeor de columnas: '))
   
for i in range(filas): 
    matriz.append([0]*columnas) #genera una matriz de 0 con dimnesiones 
                                #ingresadas anteriormente 

for i in range(filas): #Ciclo para la recoleccion de datos de la matriz 
    for j in range(columnas):
        matriz[i][j]=int(input(f'ingrese el elmento {i+1},{j+1} de la \
matriz: '))
print('La matriz original es ', matriz)
U=matriz #Se copia la matriz a una nueva variable U que se editara

#Recorrido por la matriz y sustitucion de 0
for i in range(filas): #Ciclo para el recorrido por la matriz
    for j in range(columnas): 
        if i>j: #Para una matriz superior se cumple que para la posicion i>j 
                #aparece un cero
            U[i][j]=0 #Si se cumple la condicion se coloca un cero en dicha 
                      #posicion de la matriz U
print('La matriz diagonal superior U es: ', U)

#Busqueda de 1 en las filas
contador=[0]*filas #Variable para almacenar los numero 1 contados de cada fila
for i in range(filas): #Ciclo para el recorrido de la matriz
    for j in range(columnas): 
        if U[i][j]//1 ==1: #Si en la posicion se encuentra un 1 se sumara una 
                           #unidad al contador de dicha fila
            contador[i]=contador[i]+1

parar=contador[0] #se toma el primer elemento de la variable contar 1 como 
                  #pibote en la busqeuda del mayor numero

for i in range(len(contador)):
    if contador[i]>parar:  #compara el pibote con el numero i de la lista 
                           #"contador" (donde esta almacenada la cantidad de 
                           #unos encontrada en cada fila)
        parar=contador[i]  # si es mayor al pibote actual lo sustituye hasta 
                           #acabar de recorrer toda la lista y asi encontrar 
                           #el mayor numero presente en la lista
print(f'El mayor numero de unos en una fila de U es de {parar} y se encuetra en la\
 fila numero {contador.index(parar)+1} de la matriz: ')
```
